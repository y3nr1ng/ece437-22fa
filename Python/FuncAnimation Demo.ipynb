```python
%matplotlib widget

import matplotlib.animation as anim
import matplotlib.pyplot as plt
import numpy as np
from functools import partial

fig, ax = plt.subplots(1, 1)

im = np.zeros((488, 648), np.uint8)
h = ax.imshow(im, cmap='gray')
h.set_clim(0, 255)
ax.axis('tight')

def image_update(index, im):
    if index % 16 == 0:
        im.fill(255)
    else:
        skip = 16 - index % 16 - 1
        im[:, ::skip] = 0

    h.set_data( im )
    ax.set_title(f'frame={index}')
    return h, 

ani = anim.FuncAnimation(fig, partial(image_update, im=im), interval=25, frames=None, blit=False)
plt.show()
```
